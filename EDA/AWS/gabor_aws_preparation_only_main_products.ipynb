```python
import pandas as pd
from fastavro import writer, parse_schema, reader
pd.set_option('display.max_columns', None)
```

```python
basepath = '../../datasets/preprocessed_datasets/gabor/'
customer_data = pd.read_pickle(basepath +'customer_data.pkl')
```

```python
display(customer_data.head())
```

```python
customer_data = customer_data.drop(['num_orders', 'order_ids'], axis=1).rename({'customer__id': 'USER_ID'}, axis=1)
customer_data['IS_GABOR'] = 'true'
customer_data['USER_ID'] = customer_data['USER_ID'].astype(int).astype(str)
print(customer_data)
```

```python
schema = {
    "type": "record",
    "name": "Users",
    "namespace": "com.amazonaws.personalize.schema",
    "fields": [
        {
            "name": "USER_ID",
            "type": "string"
        },
        {
            "name": "IS_GABOR",
            "type": "boolean"
        }
    ],
    "version": "1.0"
}
parsed_schema = parse_schema(schema)
```

```python
records = customer_data.to_dict('records')
display(records)
```

```python
with open(basepath + 'customer_data.avro', 'wb') as out:
    writer(out, parsed_schema, records)
customer_data.to_csv(basepath + 'customer_data.csv', index=False, header=True)
```

```python
item_data = pd.read_pickle(basepath +'only_products.pkl')
display(item_data)
```

```python
rename_cols = {'product__id': 'ITEM_ID', 'b2cPrice' :'PRICE', 'productName': 'PRODUCT_NAME', 'brand__id': 'BRAND', 'mainCategory__id':'category_l1', 'gender': 'GENDER', 'line' : 'LINE', 'material': 'MATERIAL', 'color': 'COLOR', 'productType_x': 'PRODUCT_TYPE'}

item_data = item_data.rename(rename_cols, axis=1)
item_data = item_data[rename_cols.values()]

item_data['ITEM_ID'] = item_data['ITEM_ID'].astype(str)
item_data['category_l1'] = item_data['category_l1'].astype(str)
# item_data['BRAND'] = item_data['BRAND'].astype(float)
# item_data['SIZE'] = item_data['SIZE'].astype(float)

display(item_data)
```

```python
item_schema = {
    "type": "record",
    "name": "Items",
    "namespace": "com.amazonaws.personalize.schema",
    "fields": [
        {
            "name": "ITEM_ID",
            "type": "string"
        },
        {
            "name": "PRICE",
            "type": "float"
        },
        {
            "name": "PRODUCT_NAME",
            "type": "string"
        },
        # {
        #     "name": "SIZE",
        #     "type": "float"
        # },
        {
            "name": "BRAND",
            "type": "float",
            "categorical": True
        },
        {
            "name": "category_l1",
            "type": "string",
            "categorical": True
        },
        {
        "name": "GENDER",
        "type": "string",
            "categorical": True
        },
        {
            "name": "MATERIAL",
            "type": "string",
            "categorical": True
        },
        {
            "name": "COLOR",
            "type": "string",
            "categorical": True
        },
        {
            "name": "PRODUCT_TYPE",
            "type": "string",
            "categorical": True
        },
    ],
    "version": "1.0"
}
parsed_item_schema = parse_schema(item_schema)
```

```python
item_records = item_data.to_dict('records')
display(item_records)
```

```python
with open(basepath + 'item_data.avro', 'wb') as out:
    writer(out, parsed_item_schema, item_records)

item_data.to_csv(basepath + 'item_data.csv', index=False, header=True)
```

```python
interaction_data = pd.read_pickle(basepath +'product_data.pkl')
display(interaction_data)
```
